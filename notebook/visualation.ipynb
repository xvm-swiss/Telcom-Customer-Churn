```python
import numpy as np
import pandas as pd
import os
import warnings; warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../data/preprocessing data/preprocessing_data.csv')
```

# Gender: Female, Male, Total

```python
female_sum = df['gender'].value_counts()[1]
male_sum = df['gender'].value_counts()[0]
totall_sum = df.shape[0]

print(f'Total Female: {female_sum}')
print(f'Total Male: {male_sum}')
print(f'Total: {totall_sum}')
```

```python
# Prepare data
gender_counts = df['gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

# Create bar chart
fig = px.bar(gender_counts, 
             x='Gender', 
             y='Count',
             color='Gender',
             text='Count', # Shows the number directly on the bar
             title=f'Gender Distribution (Total: {df.shape[0]})',
             color_discrete_map={'Female': 'pink', 'Male': 'blue'}) # Optional colors

fig.update_traces(textposition='auto')
# Mögliche Position
# Hier änderst du die Position: 'inside', 'outside' oder 'auto'
fig.show()
```

# Partner: No, Yes, Total

```python
Partner_no = df["Partner"].value_counts()[0]
Partner_yes = df["Partner"].value_counts()[1]
Partner_all = df.shape[0]


print(f'Partner NO: {Partner_no}')
print(f'Partner YES: {Partner_yes}')
print(f'Total Partner: {Partner_all}')
```

```python
# Prepare data
Partner_counts = df['Partner'].value_counts().reset_index()
Partner_counts.columns = ['Partner', 'Count']

# Create bar chart
fig = px.bar(Partner_counts, 
             x='Partner', 
             y='Count',
             color='Partner',
             text='Count', # Shows the number directly on the bar
             title=f'Partner (Total: {df.shape[0]})',
             color_discrete_map={'Female': 'pink', 'Male': 'blue'}) # Optional colors

fig.update_traces(textposition='auto')
# Mögliche Position
# Hier änderst du die Position: 'inside', 'outside' oder 'auto'
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 1. Erstelle das Gerüst (1 Zeile, 2 Spalten)
fig = make_subplots(rows=1, cols=2, subplot_titles=("Gender Distribution", "Partner Distribution"))

# 2. Grafik 1: Gender (Links)
fig.add_trace(
    go.Bar(x=df['gender'].value_counts().index, 
           y=df['gender'].value_counts(),
           text=df['gender'].value_counts(),
           marker_color=['blue', 'pink'], # Male/Female Farben
           name="Gender"),
    row=1, col=1
)

# 3. Grafik 2: Partner (Rechts)
fig.add_trace(
    go.Bar(x=df['Partner'].value_counts().index, 
           y=df['Partner'].value_counts(),
           text=df['Partner'].value_counts(),
           marker_color=['orange', 'green'], # Yes/No Farben
           name="Partner"),
    row=1, col=2
)

# 4. Layout & Text-Position
fig.update_traces(textposition='auto')
fig.update_layout(title_text=f"Dataset Overview (Total: {df.shape[0]})", showlegend=False)

fig.show()

```

# Partner: Yes, no | Gender: Male, Female

```python
import plotly.express as px

# 1. Daten gruppieren (statt manueller Summen)
partner_gender_data = df.groupby(['Partner', 'gender']).size().reset_index(name='Count')

# 2. Grafik erstellen: Gender auf der X-Achse, Partner als Gruppen (nebeneinander)
fig = px.bar(partner_gender_data, 
             x='gender', 
             y='Count', 
             color='Partner', 
             barmode='group',       # 'group' stellt die Balken NEBENEINANDER
             text='Count',          # Zeigt deine berechneten Summen an
             title='Partner Status by Gender,<br>Partner: No, Yes| Gender: Male, Female',
             color_discrete_map={'Yes': 'yellow', 'No': 'pink'})

# 3. Text positionieren
fig.update_traces(textposition='inside')
fig.show()
```

### Churn = Kündigung: Male, Female, Total

```python
Churn_male = df["Churn"][df["gender"] == 'Male'].value_counts().sum()
Churn_female = df["Churn"][df["gender"] == 'Female'].value_counts().sum()
Total_Churn = df["Churn"].value_counts().sum()

# 1. Daten in Listen (Nutzt deine berechneten Variablen)
labels = ['Male', 'Female']
counts = [Churn_male, Churn_female]

# 2. Grafik erstellen
fig = px.bar(x=labels, 
             y=counts, 
             color=labels,
             text=counts,
             title=f'Churn Distribution by Gender (Total: {Total_Churn})',
             labels={'x': 'Gender', 'y': 'Count'})

# 3. Styling
fig.update_traces(textposition='inside')
fig.show()
```
