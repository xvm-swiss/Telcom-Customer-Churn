```python
def encode_contract_time( df ):
    df['Contract'] = df['Contract'].map({
       'Month-to-month': 0, 'One year': 1, 'Two year': 2
    })
    return df

df = encode_contract_time (df )
```

```python
# Das Ziel (y) in Zahlen umwandeln (Löst den pos_label Fehler)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```
